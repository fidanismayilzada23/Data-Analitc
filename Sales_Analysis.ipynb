#Funnel Analysis

```python
#bu müşterilerin alış prosesinde keçdiyi merheleleri gösterir ve harada itki olduğu tapmaq üçün istifade olunur
```

```python
purchase= df[df['Sales']>0]['Customer ID'].unique() #sales
checkout=df[df['Profit']>0]['Customer ID'].unique() #checkout
cart=df[df['Discount']>0]['Customer ID'].unique() #cart

funnel={
    'Visited Product Page': len(df['Customer ID'].unique()), #Her müşteri bir defa sayılsın
    'Added to Cart': len(set(cart)),
    'Checkout Started': len(set(cart) & set(checkout)),
    'Purchase Completed': len(set(cart) & set(checkout) & set(purchase))
}
#Data frame şekline düşsün
funnel_df= pd.DataFrame.from_dict(funnel, orient='index', columns=['Users'])

#Drop-off faizi
#Drop-off=1-(Next step users / current step users)
funnel_df['Drop-off %']=1-funnel_df['Users'].shift(-1)/funnel_df['Users']
funnel_df['Drop-off %']=funnel_df['Drop-off %'].fillna(0)*100

print(funnel_df)
```

#Cohort analys

```python
#daimi müşterileri analiz etmek için
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])

#İlk alış ayın
df['CohortMonth']=df.groupby('Customer ID')['Order Date'].transform('min').dt.to_period('M')
#sifariş ayı
df['OrderMonth']=df['Order Date'].dt.to_period('M')

#Cohort index : neçe ay keçib

df['CohortIndex']=((df['OrderMonth'].dt.year-df['CohortMonth'].dt.year)*12+
                   (df['OrderMonth'].dt.month-df['CohortMonth'].dt.month)+1)
#Pivot tabel
cohort_data=df.groupby(['CohortMonth','CohortIndex']).nunique().reset_index()
cohort_counts=cohort_data.pivot(index='CohortMonth',columns='CohortIndex',values='Customer ID')

cohort_size=cohort_counts.iloc[:,0]
retention=cohort_counts.divide(cohort_size, axis=0)
print(retention)
```

```python
#df_cohort
plt.figure(figsize=(20,8))
sns.heatmap(
    retention,
    annot=True, #hüceyrede değeri gösterir
    fmt=".2f", #2 onluq deyer şeklinde gösterir
    cmap="YlGnBu", #reng sxemi
    cbar=True)
plt.title("Cohort Analysis - Customer Retention",fontsize=18)
plt.xlabel("CohortIndex (Months After First Purchase)")
plt.ylabel("Cohort Month (First Purchase)")
plt.show()
```

#RFM

```python
#Müşterileri davranışlarına göre seqmentleşdirmek üçün istifade olunan metotdur

#R (Recency) -  Müşteri ne qeder yaxın zamanda alıb?

#F (Frequency)- Neçe defe alıb?

#M (Monetary)- Ne qeder pul xerclenib?
# tez- tez yaxın zamanda ve çox pul xercleyen müşteriler. En deyerli müşterini tapırıq
```

```python
snapshot_date=df['Order Date'].max() + pd.Timedelta(days=1)

rfm=df.groupby ('Customer ID').agg({
    'Order Date':lambda x: (snapshot_date - x.max()).days, #Recency
    'Order ID': 'nunique', #Frequency
    'Sales':'sum' #Monetary
}).reset_index()
rfm.columns=['Customer ID','Rececy','Frequency','Monetary']
```

```python
rfm
```

```python
rfm['R_Score']=pd.qcut(rfm['Rececy'],5,labels=[5,4,3,2,1])
rfm['F_Score']=pd.qcut(rfm['Frequency'].rank(method='first'),5,labels=[1,2,3,4,5])
rfm['M_Score']=pd.qcut(rfm['Monetary'],5,labels=[1,2,3,4,5])
```

```python
rfm
```

```python
rfm['RFM_Segment']=(
    rfm['R_Score'].astype(str)+
    rfm['F_Score'].astype(str)+
    rfm['M_Score'].astype(str)
)
rfm
```

```python
rfm[['R_Score','F_Score','M_Score']]=rfm[['R_Score','F_Score','M_Score']].astype(int)
```

```python
#Champions -> en deyerli müşterliler

#Loyal -> aktiv ve stabil

#At Risk -> Gede bilecek müşteriler

#Others -> normal
```

```python
def segment(row):
  if row['R_Score']==5 and row['F_Score']>=4:
      return 'Champions'
  elif row['R_Score']>=4:
      return 'Loyal'
  elif row['R_Score']<=2 and row['F_Score']>=3:
      return 'At Risk'
  else:
      return 'Others'
rfm['Segment']=rfm.apply(segment,axis=1)
```

RFM + Churn elaqesi

```python
snapshot_date=df['Order Date'].max() + pd.Timedelta(days=1)

last_purchase=df.groupby('Customer ID')['Order Date'].max().reset_index()
last_purchase['days_since_last']=(snapshot_date-last_purchase['Order Date']).dt.days
last_purchase['churned']=last_purchase['days_since_last']>90
```

```python
rfm_churn = rfm.merge(
    last_purchase[['Customer ID','churned']],
    on='Customer ID',
    how='left'
)
rfm_churn
```

```python
churn_table=pd.crosstab(
    rfm_churn['Segment'],
    rfm_churn['churned'],
    normalize='index'
) * 100
churn_table
```

```python
import pandas as pd
import numpy as np

#others ve AT Risk segmentinden churn olmuş müşteriler
priority_customers=rfm_churn[
    (rfm_churn['Segment'].isin(['Others','At Risk']))&
    (rfm_churn['churned']==True)
].copy()

#Monetary + Frequency esasında prioritetledir
priority_customers['Priority_Score']=priority_customers['Monetary']*priority_customers['Frequency']

#Priority sıralama
priority_customers=priority_customers.sort_values(by='Priority_Score',ascending=False)

#Yalnız lazım olan sütunlar
priority_list=priority_customers[['Customer ID','Segment','Monetary','Frequency','Priority_Score']]
priority_list.head(10)
```

```python
#CLV (Customer Lifetime Value) bir müşterinin şirket üçün gelecek
```

```python
rfm_churn['AOV']=rfm_churn['Monetary'] / rfm_churn['Frequency']
rfm_churn ['Retention'] = (~rfm_churn ['churned']).astype(int)
rfm_churn['CLV'] = rfm_churn ['AOV'] * rfm_churn ['Frequency'] * rfm_churn['Retention']
# Segment üzrə ortalama CLV
segment_clv = rfm_churn.groupby('Segment')['CLV'].mean().sort_values (ascending=False)
segment_clv
```

#A/B TEST

```python
#Other müşterileri xilas etmek
```

```python
import numpy as np
others = rfm_churn[rfm_churn['Segment'] == 'Others'].copy()
np.random.seed(42)
others['group'] = np.random.choice(['A', 'B'], size=len(others))
```

```python
np.random.seed(42)
others['converted'] = np.where(
others['group'] == 'B',
np.random.binomial (1, 0.30, len (others)),
np.random.binomial (1, 0.20, len(others))
)
```

```python
conversion_rates=others.groupby('group')['converted'].mean()
conversion_rates
```

```python
from statsmodels.stats.proportion import proportions_ztest
success = others.groupby('group')['converted'].sum()
nobs = others.groupby('group')['converted'].count()
z_stat, p_value = proportions_ztest(success, nobs)
p_value
```
