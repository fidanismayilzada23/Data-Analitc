```python
cars=sns.load_dataset('mpg')
cars;
```

```python
sns.get_dataset_names()
```

```python
cars.shape
```

```python
cars.info()
# null deÄer var
```

```python
cars.isnull().sum()
```

```python
cars.dropna(inplace=True)
```

```python
cars.isnull().sum()
```

```python
sns.relplot(x='model_year',y='mpg',col='origin',hue='cylinders',data=cars)
```

```python
sns.kdeplot(cars.horsepower,fill=True)#kernel destination
```

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg,levels=15,fill=True)
```

```python
sns.histplot(cars.horsepower,kde=True,stat='density')
```

```python
sns.histplot(x='horsepower',data=cars,hue='origin',multiple='fill')
```

```python
sns.histplot(x='model_year',hue='origin',data=cars,multiple='dodge',shrink=0.8)
```

```python
sns.pairplot(data=cars,vars=['model_year','horsepower','weight'],hue='origin')
```

```python
sns.boxplot(x='origin',y='mpg',data=cars)
```

```python
sns.violinplot(x='origin',y='mpg',data=cars)
```

```python
sns.jointplot(x='horsepower',y='mpg',data=cars)
```

```python
corr = cars.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
sns.countplot(x='origin',data=cars)
```

```python
sns.lmplot(x='weight',y='mpg',hue='origin',data=cars)
```

```python
sns.swarmplot(x='origin',y='mpg',data=cars)
```

```python
sns.barplot(x='origin',y='mpg',data=cars,estimator='median')
```

```python
sns.scatterplot(x='horsepower',y='mpg',data=cars)
```

```python
sns.jointplot(x='horsepower',y='mpg',data=cars,kind='kde')
```

```python
sns.lineplot(x='model_year',y='mpg',data=cars,hue='origin')
```

```python
sns.stripplot(x='origin',y='mpg',data=cars) #nÃ¶qteler Ã¼st Ã¼ste dÃ¼ÅmÃ¼r outlierleri rahat gÃ¶rÃ¼rÃ¼k
```

```python
sns.jointplot(x='horsepower',y='mpg',data=cars,kind='hex',gridsize=20,cmap='Blues')
```
