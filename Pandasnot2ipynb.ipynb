#Missing Data

```python
dfm=pd.read_csv('/content/car-sales-missing-data.csv')
dfm.head()
```

```python
#Setir stun sayn gsterir.
dfm.shape
```

```python
dfm.info()
```

```python
# Her stunda null obyektlerin sayn gsterir.
dfm.isnull().sum()
```

```python
dfm
```

```python
#Bo deyerleri silmek
dfm=dfm.dropna(subset='Price')
dfm
```

```python
dfm.isnull().sum()
```

```python
#Mean ile boluqlar doldurmaq.
dfm['Odometer'].fillna(dfm['Odometer'].mean(),inplace=True)
dfm
```

```python
#Text tipilileri ise moda ile doldururuq
dfm['Colour']=dfm['Colour'].fillna(dfm['Colour'].mode()[0])
dfm
```

```python
for col in ['Make','Colour']:
  dfm[col]=dfm[col].fillna(dfm[col].mode()[0])
dfm.isna().sum()
```

#Pandas Merge

```python
table_a=pd.DataFrame({'id':['a1','a2','a3','a4'],'Value':['A','B','C','AA']})
table_b=pd.DataFrame({'id':['a1','a2','a3','a5'],'Value':['D','E','F','BB']})

pd.merge(table_a,table_b,on='id')
```

```python
table_a=pd.DataFrame({'id':['a1','a2','a3','a4'],'Value':['A','B','C','AA']})
table_b=pd.DataFrame({'id':['a1','a2','a3','a5'],'Value':['D','E','F','BB']})

pd.merge(table_a,table_b,on='id',how='left')
#left join etmek istediyimiz zaman
```

```python
table_a=pd.DataFrame({'id_1':['a1','a2','a3','a4'],'Value':['A','B','C','AA']})
table_b=pd.DataFrame({'id_2':['a1','a2','a3','a5'],'Value':['D','E','F','BB']})
```

```python
#Stun ad eyni olmad zaman bu ekilde birlediririk.
pd.merge(table_a,table_b,left_on='id_1',right_on='id_2')
```
